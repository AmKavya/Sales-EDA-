# EDA

### Gender 

```python
df.columns
```

we can see female has more purchase than male 

```python
ax = sns.countplot(x='Gender',data = df,hue = 'Gender' , palette = 'pastel' , legend = True)
for container in ax.containers:
    ax.bar_label(container)
# ax.bar_label(ax.containers[0])
```

```python
# same but piechart 
counts = df['Gender'].value_counts()
plt.pie(
    counts.values,
    labels = counts.index,
    startangle = 90,
    autopct = '%.1f%%',
    colors = sns.color_palette('Paired')

)
# plt.axis('equal')  # to get perfect circle 
plt.show()
```

We can see females contribute most to frequency as well as most to revenue 

```python
# How much amount spent by gender 
sales_gen  = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)
sale = sns.barplot(x='Gender', y = 'Amount',data = sales_gen,hue = 'Gender',palette = 'pastel',legend = True)
sale.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x , pos:f'{x/1e6:.1f}M')) # for y axis formating 
for container in sale.containers:
    sale.bar_label(container,fmt = lambda x : f'{x/1e6:.1f}M')  #fmt = %.0f,%.1f
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum()
plt.pie(
    sales_gen['Amount'],
    labels = sales_gen['Gender'], # as this gives out dataframes and not series to have value index hence this 
    autopct = '%.1f%%',
    startangle = 90 ,
    colors = sns.color_palette('Paired')

)
plt.show()
```

### age


```python
df.columns
```

```python
age_order = df['Age Group'].value_counts().sort_values(ascending = False).index # to sort bars in descending order based on counts  
age = sns.countplot(x = 'Age Group',data = df, hue = 'Gender',palette  = 'pastel',order = age_order) # .index as order requires labels that are in axis 
for container in age.containers: 
    age.bar_label(container)
```

```python
# Total Amount vs age groups
sales_age = df.groupby(['Age Group','Gender'],as_index = False)["Amount"].sum().sort_values(by='Amount',ascending = False)
age2 = sns.barplot(x='Age Group', y ='Amount', data = sales_age,hue='Gender',palette = 'pastel')
age2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x , pos:f'{x/1e6:.1f}M'))
for container in age2.containers:
    age2.bar_label(container,fmt = lambda x : f'{x/1e6:.1f}M')
```

most spenders n most buyers are female in age group 26 - 35

```python
# total amount of top 10 states by orders 
states2 = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)

#color mapping 
norm = plt.Normalize(vmin = states2['Amount'].min(),vmax=states2['Amount'].max())
cmap = plt.cm.Blues
colors =cmap(norm(states2['Amount'])).tolist()

#plotting
state2 = sns.barplot(y = 'State',x='Amount',data = states2,hue ='State',palette = colors)

#Labels axis format
state2.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x , pos:f'{x/1e6:.1f}M'))
sns.despine(right = True) 

#labels bar
for container in state2.containers:
    state2.bar_label(container,fmt = lambda x : f'{x/1e6:.1f}M')

# color bar
cbar = plt.colorbar(plt.cm.ScalarMappable(norm=norm,cmap=cmap),orientation = 'horizontal' ,ax=state2,label='Amount')
#label formatting of colorbar 
cbar.ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x , pos:f'{x/1e6:.1f}M'))
```
